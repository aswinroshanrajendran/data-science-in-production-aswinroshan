```python
import numpy as np
import matplotlib.pyplot as plt

# Create a NumPy array with shape (10, 3) and fill it with random integers
random_array = np.random.randint(low=1, high=100, size=(10, 3))

# Visualize the array using a scatter plot in 2D
plt.scatter(random_array[:, 0], random_array[:, 1], label='Random Points')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Random 2D Points')

plt.legend()
plt.show()
```
